```python
import cv2
import numpy as np
from PIL import Image, ImageEnhance, ImageOps
import matplotlib.pyplot as plt

# Load the image
image_path = 'Case 2.jpg'
original_image = Image.open(image_path)

# Enhance the image brightness and contrast
enhancer = ImageEnhance.Brightness(original_image)
brightened_image = enhancer.enhance(1.5)  # Increase brightness

enhancer = ImageEnhance.Contrast(brightened_image)
enhanced_image = enhancer.enhance(2.0)  # Increase contrast

# Convert the enhanced image to OpenCV format for further processing
enhanced_image_cv = cv2.cvtColor(np.array(enhanced_image), cv2.COLOR_RGB2BGR)

# Add border around the photo
photo_border_color = (255, 0, 0)  # Red color
photo_coords = (50, 275, 325, 625)  # Adjust the coordinates as needed
# cv.rectangle(photo, pt1=(50,275), pt2=(325,625), color=(0,255,0), thickness=2)

cv2.rectangle(enhanced_image_cv, (photo_coords[0], photo_coords[1]), (photo_coords[2], photo_coords[3]), photo_border_color, 5)

# Add border around the name
name_border_color = (255, 0, 0)  # Red color
name_coords = (510, 320, 1035, 365)  # Adjust the coordinates as needed

cv2.rectangle(enhanced_image_cv, (name_coords[0], name_coords[1]), (name_coords[2], name_coords[3]), name_border_color, 5)

# Convert back to PIL format for displaying
enhanced_image_with_borders = Image.fromarray(cv2.cvtColor(enhanced_image_cv, cv2.COLOR_BGR2RGB))

# Convert the enhanced image to grayscale
enhanced_image_gray = ImageOps.grayscale(enhanced_image_with_borders)

# Display the original, enhanced RGB, and enhanced grayscale images
plt.figure(figsize=(20, 10))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(original_image)

plt.subplot(1, 3, 2)
plt.title('Enhanced RGB Image')
plt.imshow(enhanced_image_with_borders)

plt.subplot(1, 3, 3)
plt.title('Enhanced Grayscale Image')
plt.imshow(enhanced_image_gray, cmap='gray')

plt.show()
```

```python
original = cv.imread('Case 2.jpg')
original = cv.cvtColor(original, cv.COLOR_BGR2RGB)
contrass_image = np.zeros(original.shape, original.dtype)

# akses per pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      contrass_image[y, x, c] = np.clip(4 * original[y, x, c] + 3.5, 0, 255)

image_rgb = np.array(original)

# Averaging method
grayscale_avg = np.mean(image_rgb, axis=2)

plt.figure(figsize=(20, 10))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(original)

plt.subplot(1, 3, 2)
plt.title('Enhanced RGB Image')
plt.imshow(contrass_image)

plt.subplot(1, 3, 3)
plt.title('Enhanced Grayscale Image')
plt.imshow(grayscale_avg, cmap='gray')
```

```python
original = cv.imread('Case 2.jpg')
original = cv.cvtColor(original, cv.COLOR_BGR2RGB)
contrass_image = np.zeros(original.shape, original.dtype)

# akses per pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      contrass_image[y, x, c] = np.clip(4 * original[y, x, c] + 3.5, 0, 255)

image_rgb = np.array(original)

# Averaging method
grayscale_avg = np.mean(image_rgb, axis=2)

cv.rectangle(contrass_image, pt1=(50,275), pt2=(325,625), color=(0,255,0), thickness=2)
cv.rectangle(contrass_image, pt1=(510,320), pt2=(1035,365), color=(0,255,0), thickness=2)

cv.rectangle(grayscale_avg, pt1=(50,275), pt2=(325,625), color=(0,255,0), thickness=2)
cv.rectangle(grayscale_avg, pt1=(510,320), pt2=(1035,365), color=(0,255,0), thickness=2)

plt.figure(figsize=(20, 10))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(original)

plt.subplot(1, 3, 2)
plt.title('Enhanced RGB Image')
plt.imshow(contrass_image)

plt.subplot(1, 3, 3)
plt.title('Enhanced Grayscale Image')
plt.imshow(grayscale_avg, cmap='gray')
```
